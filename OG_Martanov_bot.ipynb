```python
#функция по отбору словарей(коктейлей) по заданным параметрам(основа, вкус)

def selec(base: str, sheet: list):
    selected = []
    for i in range(0,len(sheet)):
        for value in sheet[i].values():
            if base == value:
                selected.append(sheet[i])
                continue
    return selected
                
```
